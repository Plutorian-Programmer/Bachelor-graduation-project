```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
map_path = "C:/Users/karst/OneDrive/Documenten/uni/jaar 3 semester 2, het semester dat ik opgaf/Bachelor-graduation-project/code/notebooks/"
```

```python
with open(map_path+'performance_dict_new.pkl', 'rb') as f:
    performance_dict_09 = pickle.load(f)
```

```python
temp_plot_list = []

for key in performance_dict_09:
    temp_plot_list.append(performance_dict_09[key]["avg_recall"])
```

```python
fig = plt.figure(figsize=(10,10), facecolor="gray")
ax = fig.add_axes([0,0,1,1])
labels = [i/10 for i in range(1,10)]
ax.violinplot(temp_plot_list, showmeans=True, widths=0.75)
ax.plot([1,2,3,4,5,6,7,8,9],np.mean(temp_plot_list, axis=1))
ax.set_xticks([1,2,3,4,5,6,7,8,9])
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)
ax.set_xticklabels(labels)
ax.set_ylim(0.15,0.5)
ax.set_xlabel(xlabel="percentage of data which is used as trainings data", fontsize=18)
ax.set_ylabel(ylabel="recall",fontsize=18)
ax.legend(["mean of recall","distribution of recall"], fontsize=12)
ax.set_title("Distribution of recall over data splits with a 10% test split", fontsize=20)
plt.show()
```
